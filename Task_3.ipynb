Language Model Analysis Project (Advanced Level)

This project explores the capabilities of a language model (GPT-2) in generating coherent, creative, and contextually relevant text

```python
!pip install transformers wordcloud matplotlib

from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import torch
import random

torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Load GPT-2 Model and Tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Prompt Experiments
prompts = [
    "In the future, artificial intelligence will",
    "The main challenge of machine learning is",
    "Once upon a time in a galaxy far away",
    "A healthy lifestyle consists of",
    "The impact of climate change is"
]

outputs = []
for prompt in prompts:
    result = generator(prompt, max_length=60, num_return_sequences=1, temperature=0.7)[0]['generated_text']
    print(f"\nPrompt: {prompt}\nOutput: {result}\n{'-'*80}")
    outputs.append(result)

# Qualitative Analysis / Research Questions
# Visualization of Results

# 1. Word Cloud
combined_text = " ".join(outputs)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Generated Text")
plt.show()

# 2. Length vs Coherence (hypothetical subjective scoring)
lengths = [30, 60, 90, 120]
coherence_scores = [0.7, 0.75, 0.6, 0.5]
plt.figure(figsize=(8, 5))
plt.plot(lengths, coherence_scores, marker='o')
plt.title("Length vs Subjective Coherence")
plt.xlabel("Max Token Length")
plt.ylabel("Coherence Score")
plt.grid(True)
plt.show()

# 3. Response Diversity by Prompt
response_lengths = [len(tokenizer.encode(text)) for text in outputs]
plt.figure(figsize=(8, 5))
plt.barh(prompts, response_lengths, color='skyblue')
plt.xlabel("Token Count")
plt.title("Response Length per Prompt")
plt.show()

# Best Practices & Ethical Considerations
# - Bias Detection: Check outputs for stereotypical or biased responses.
# - Safety: Filter inappropriate or hallucinated content.
# - Transparency: Declare limitations of the model.

# Conclusion and Future Work
print("""
Conclusions:
- GPT-2 shows strong contextual and grammatical skills.
- Creativity is high, but factual correctness can vary.
- Long outputs sometimes lose coherence.

Future Work:
- Try instruction-tuned models like T5 or FLAN-T5.
- Fine-tune GPT-2 on a specific domain.
- Use BLEU or ROUGE scores for summarization evaluation.
""")
```
